```python
import numpy as np
from sklearn.cluster import KMeans # 聚类算法模块的kmeans算法
from sklearn import metrics# scikit-learn中的度量评价指标模块
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(8,10))
plt.subplot(3,2,1)
x1 = np.array([1, 2, 3, 1, 5, 6, 5, 5, 6, 7, 8, 9, 7, 9])
x2 = np.array([1, 3, 2, 2, 8, 6, 7, 6, 7, 1, 2, 1, 1, 3])
X = np.array(list(zip(x1,x2))).reshape(len(x1),2)
plt.xlim([0,10])
plt.ylim([0,10])
plt.title('Sample')
plt.scatter(x1,x2)  # 通过散点图查看一下点的大致分布
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'b']
markers = ['o', 's', 'D', 'v', '^', 'p', '*', '+']
tests = [2, 3, 4, 5, 8]
subplot_counter = 1

for t in tests: # 选取tests中数据作为k点选项
    subplot_counter += 1
    plt.subplot(3,2,subplot_counter)
    kmeans_model = KMeans(n_clusters=t).fit(X) # 直接一步生成kmeans分类
    for i,l in enumerate(kmeans_model.labels_):
        plt.plot(x1[i],x2[i],color=colors[l],marker=markers[l],ls='None') #ls='None' 是代表不显示线条
        plt.xlim([0, 10])
        plt.ylim([0, 10])
        # metrics.silhouette_score 求轮廓系数 通过欧式距离
        plt.title('K = %s, SCoefficient = %.03f' % (t, metrics.silhouette_score(X, kmeans_model.labels_,metric='euclidean')))
    
```
